```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 

# Sample data dictionary
data_dic = {
    'A': [12, 15, 45, 71],
    'B': [18, 26, 50, 80],
    'C': [32, 48, 78, 93],
    'D': [29, 35, 63, 81]
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data_dic, index=[0, 1, 2, 3])

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap="magma", fmt="d")
plt.title("Heatmap of Data")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```
