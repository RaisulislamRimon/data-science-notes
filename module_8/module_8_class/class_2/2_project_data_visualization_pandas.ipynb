### Using Pyplot

```python
plt.style.use('dark_background')
```

```python
plt.subplots(figsize=(10,5))
plt.plot(gas['Year'], gas['USA'], label='USA', marker='o', markeredgecolor="red")
plt.plot(gas['Year'], gas['Canada'], label='Canada', marker='o', markeredgecolor="red")
plt.xticks(gas.Year[::3].tolist())
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18, 'color':'g'})

plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas['Canada'], 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='SKorea')
plt.plot(gas.Year, gas.Australia, 'y.-', label='Austrolia')

plt.xticks(gas.Year[::3].tolist()) #[1990, 1993, 1996, 1999, 2002, 2005, 2008]

plt.xlabel('Year')
plt.ylabel('US Dollars')

#plt.grid()

plt.legend()

plt.savefig('Gas_price_figure.png', dpi=300, bbox_inches ='tight')

#plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot()


plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18, 'color':'g'})

plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas['Canada'], 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='SKorea')
plt.plot(gas.Year, gas.Australia, 'y.-', label='Austrolia')

plt.xticks(gas.Year[::3].tolist()) #[1990, 1993, 1996, 1999, 2002, 2005, 2008]

for i,j in zip(gas.Year,gas['South Korea']):
    if i in gas.Year[::3].tolist():
        ax.annotate('%s' %j, xy=(i,j), xytext=(0,10), textcoords='offset points')
    
plt.show()
```

## Pie Plot
We can create a pie plot with  `Series.plot.pie()` or `DataFrame.plot.pie()`. If your data includes any NaN, they will be automatically filled with 0. A ValueError will be raised if there are any negative values in your data.<br>
Let's try pie plot with df2.<br>

We need to specify a target column by the y argument. When y is specified, pie plot of selected column will be drawn.
