Question: 
Make the following Dataframe by using pandas. Write the functions to see the how many NaN values are in the entire Dataframe? Clean the NaN values using Forward fill, Backward fill, ‘0’ Fill

```python
# import the necessary libraries
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Make the following Dataframe by using pandas
# creating the dataFrame
data = {
    'A': [21, np.nan, 22, 4, np.nan],
    'B': [31, 32, 33, 34, 35],
    'C': [np.nan, np.nan, np.nan, np.nan, np.nan],
    'D': [46, 47, np.nan, 48, 49]
}
df = pd.DataFrame(data)
df
```

```python
# Write the functions to see the how many NaN values are in the entire Dataframe?
total_nan = df.isna().sum().sum()
print("Total NaN values in the entire DataFrame: ", total_nan)
column_wise_nan = df.isna().sum()
print("column_wise_nan:\n", column_wise_nan)
```

```python
df
```

 Clean the NaN values using Forward fill, Backward fill, ‘0’ Fill

```python
#  Clean the NaN values using Forward fill
df.fillna(method='ffill')
```

```python
#  Clean the NaN values using Backward fill
df.fillna(method='bfill')
```

```python
# show the dataframe
df
```

```python
#  Clean the NaN values using ‘0’ Fill
df.fillna(0)
```
