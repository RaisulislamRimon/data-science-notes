Write the code for implementation KNN. (No need scaling and train test split. Just separate the features and target and write the code to make a model using KNN algorithm)

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('data.csv')
df 
```

```python
# Separate features (C1 to C10) and target (Result)
X = df[["C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9", "C10"]]
y = df["Result"]
```

```python
# Initialize the KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Fit the model
knn.fit(X, y)
```

```python
# Make a prediction
data_for_predict = [[50, 60, 30, 45, 70, 30, 40, 50, 60, 70]]  # Replace with your sample
prediction = knn.predict(data_for_predict)
```

```python
print(f"Prediction for the sample data: {prediction}")
```
