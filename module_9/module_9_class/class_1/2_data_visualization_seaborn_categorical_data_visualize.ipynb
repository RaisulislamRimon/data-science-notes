barplot()

```python
tips[tips['day'] == 'Thur']['total_bill'].mean()
```

```python
sns.barplot(x='day', y='total_bill', data=tips) # it will show mean of total_bill by default
```

```python
# sns.barplot(x='day', y='total_bill', data=tips) # it will show mean of total_bill by default
datalabel = sns.barplot(x='day', y='total_bill', data=tips, estimator=np.sum) # it will show mean of total_bill by default

for i in datalabel.containers:
    datalabel.bar_label(i)
```

```python
import numpy as np 
datalabel = sns.barplot(x='day', y='total_bill', data=tips, estimator=np.sum)
```

```python
import numpy as np 
datalabel = sns.barplot(x='day', y='total_bill', data=tips, estimator=np.sum)

for i in datalabel.containers:
    datalabel.bar_label(i)
```

```python
import numpy as np 
datalabel = sns.barplot(x='day', y='total_bill', data=tips, estimator=np.sum, ci=None)

for i in datalabel.containers:
    datalabel.bar_label(i)
```
