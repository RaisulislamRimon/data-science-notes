```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    'key': ['a', 'b', 'c', 'd', 'e'],
    'A1': range(5),
    'B1': range(5, 10)
})
```

```python
df2 = pd.DataFrame({
    'key': ['a', 'b', 'c'],
    'A2': range(3),
    'B2': range(3, 6)
})
```

```python
# inner join
pd.merge(df1, df2, how='inner', on='key')
```

```python
# inner join
pd.merge(df1, df2, how='inner', on='key')[['key', 'A1', 'B1']]
```

```python
# inner join
pd.merge(df1, df2, how='inner', on='key')[['key', 'A2', 'B2']]
```

```python
# left join
pd.merge(df1, df2, how='left', on='key')
```

```python
# right join
pd.merge(df1, df2, how='right', on='key')
```

```python
pd.merge(df1, df2, how='right', on='key')[['A1', 'B2']]
```

```python
# full join/outer join
pd.merge(df1, df2, how='outer', on='key')
```

```python
# Definition of the first DataFrame
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}, index=[0, 1, 2, 3])

# Definition of the second DataFrame
df2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
}, index=[4, 5, 6, 7])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2]) # default axis is 0/row
```

```python
pd.concat([df1, df2], axis=0) # default axis is 0/row
```

```python
pd.concat([df1, df2], axis=1) # axis 1 is column
```
