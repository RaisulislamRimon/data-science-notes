Above data stored in a Dataframe named df . 
Write the code for one hot encoding of Car Model.

```python
import pandas as pd

df = pd.read_csv('car_data.csv')
```

```python
df
```

```python
# Perform one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Car Model'], prefix='CarModel')

# Display the encoded DataFrame
print(df_encoded)
```
