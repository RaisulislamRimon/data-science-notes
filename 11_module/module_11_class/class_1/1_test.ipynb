```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create the dataset
data = pd.DataFrame({
    "name": ["mohan", "maria", "sakib", "tao", "virat", "khusbu", "dmitry", "selena", "john", "imran",
             "jose", "deepika", "yoseph", "binod", "gulshan", "johnson", "donald", "aamir", "ken", "Liu"],
    "height": [1.2, 2.3, 4.9, 5.1, 5.2, 5.4, 5.5, 5.5, 5.6, 5.6, 5.8, 5.9, 6.0, 6.1, 6.2, 6.5, 7.1, 14.5, 23.2, 40.2]
})

# Calculate Q1, Q3, and IQR
Q1 = np.percentile(data['height'], 25)
Q3 = np.percentile(data['height'], 75)
IQR = Q3 - Q1

# Calculate lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = data[(data['height'] < lower_bound) | (data['height'] > upper_bound)]

# Set plot style
sns.set(style="whitegrid")

# Create a figure with multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Boxplot
sns.boxplot(data=data, x='height', ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title("Boxplot of Heights")
axes[0, 0].axvline(lower_bound, color='red', linestyle='--', label='Lower Bound')
axes[0, 0].axvline(upper_bound, color='green', linestyle='--', label='Upper Bound')
axes[0, 0].legend()

# 2. Histogram
sns.histplot(data['height'], bins=15, kde=True, ax=axes[0, 1], color='orange')
axes[0, 1].set_title("Histogram of Heights")
axes[0, 1].axvline(lower_bound, color='red', linestyle='--', label='Lower Bound')
axes[0, 1].axvline(upper_bound, color='green', linestyle='--', label='Upper Bound')
axes[0, 1].legend()

# 3. Scatter plot
axes[1, 0].scatter(data.index, data['height'], color='blue', label='Data Points')
axes[1, 0].axhline(lower_bound, color='red', linestyle='--', label='Lower Bound')
axes[1, 0].axhline(upper_bound, color='green', linestyle='--', label='Upper Bound')
axes[1, 0].set_title("Scatter Plot of Heights")
axes[1, 0].set_xlabel("Index")
axes[1, 0].set_ylabel("Height")
axes[1, 0].legend()

# 4. Strip plot
sns.stripplot(data=data, x='height', ax=axes[1, 1], color='purple', size=7)
axes[1, 1].set_title("Strip Plot of Heights")
axes[1, 1].axvline(lower_bound, color='red', linestyle='--', label='Lower Bound')
axes[1, 1].axvline(upper_bound, color='green', linestyle='--', label='Upper Bound')
axes[1, 1].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
